````python
import math

# Initial list of available product names
AVAILABLE_PRODUCTS = ["Laptop", "Mouse", "Keyboard", "Monitor", "Webcam"]

## 1. Inventory Management Functions with Comprehension

def initialize_inventory(products):
    """
    Initializes the inventory using a DICTIONARY COMPREHENSION 
    and user input for initial quantities.
    """
    print("\n--- Inventory Initialization ---")
    
    # Dictionary Comprehension for initialization
    inventory = {
        product: int(input(f"Enter the quantity of '{product}' available: ")) 
        for product in products
    }
    return inventory

def update_inventory(customer_orders, inventory):
    """
    Updates the inventory based on customer orders and uses a DICTIONARY 
    COMPREHENSION to filter out products with zero quantity.
    """
    print("\n--- Updating Inventory ---")
    
    # 1. Decrement quantities for ordered products
    for product in customer_orders:
        if product in inventory and inventory[product] > 0:
            inventory[product] -= 1
            print(f"Inventory reduced for '{product}'. New quantity: {inventory[product]}")
        elif product in inventory and inventory[product] == 0:
            print(f"Warning: '{product}' was ordered but is out of stock.")
            
    # 2. Dictionary Comprehension to filter out zero-quantity items
    # This creates a NEW dictionary, effectively removing items with quantity <= 0
    updated_inventory = {
        product: quantity
        for product, quantity in inventory.items() 
        if quantity > 0
    }
    
    return updated_inventory

def print_updated_inventory(inventory):
    """
    Prints the final state of the inventory dictionary, only showing items 
    with positive stock. Uses a concise loop for clean output.
    """
    print("\n--- Final Updated Inventory ---")
    for product, quantity in inventory.items():
        print(f"**{product}**: {quantity}")

## 2. Order Management Functions

def get_customer_orders():
    """
    Prompts the user for the number of orders and gathers product names 
    using a list/set comprehension-style approach with user input.
    
    Note: Since the input function must be called inside the loop's 
    context, we use a standard loop to collect names, and then convert 
    to a set for uniqueness.
    """
    print("\n--- Customer Order Entry ---")
    
    while True:
        try:
            num_orders = int(input("Enter the number of unique products to order: "))
            if num_orders >= 0:
                break
            else:
                print("Please enter a non-negative number.")
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    # Standard loop to collect input, then a SET to ensure unique products
    ordered_products = set()
    for i in range(num_orders):
        product_name = input(f"Enter the name of product #{i+1} to order: ").strip()
        ordered_products.add(product_name)
    
    # Note: If we were gathering *quantities* with a comprehension, 
    # the syntax would be cleaner, but here, the input() call must be inside 
    # the loop, which makes a standard approach clearer for sequential inputs.
    
    # Return the set of unique orders
    return ordered_products

def calculate_order_total_price(customer_orders):
    """
    Prompts the user for the price of each ordered product and calculates 
    the total using a GENERATOR EXPRESSION inside the sum() function.
    
    :param customer_orders: The set of unique products ordered.
    :return: The total cost of the order (float).
    """
    print("\n--- Calculating Total Price ---")
    
    # Generator Expression for calculating the sum of prices
    total_price = sum(
        float(input(f"Enter the price of '{product}': "))
        for product in customer_orders
    )
    
    return total_price

## 3. Statistics Functions (From previous exercise, kept for completeness)

def calculate_order_statistics(customer_orders, products):
    """Calculates order statistics."""
    total_products_ordered = len(customer_orders)
    total_available_products = len(products)
    
    if total_available_products > 0:
        percentage_ordered = (total_products_ordered / total_available_products) * 100
    else:
        percentage_ordered = 0.0
        
    return total_products_ordered, percentage_ordered

def print_order_statistics(order_statistics):
    """Prints the calculated order statistics."""
    total_ordered, percentage = order_statistics
    
    print("\n--- Order Statistics ---")
    print(f"Total Products Ordered: {total_ordered}")
    print(f"Percentage of Unique Products Ordered: {percentage:.1f}%")

## --- Main Program Execution ---

# 1. Initialize Inventory (uses comprehension)
inventory_data = initialize_inventory(AVAILABLE_PRODUCTS)

# 2. Get Customer Orders (uses loop/set for unique product names)
orders_set = get_customer_orders()

# 3. Calculate Statistics
stats = calculate_order_statistics(orders_set, AVAILABLE_PRODUCTS)
print_order_statistics(stats)

# 4. Update Inventory (uses comprehension to filter out zero stock)
final_inventory = update_inventory(orders_set, inventory_data)
print_updated_inventory(final_inventory)

# 5. Calculate and Print Total Price (uses generator expression)
total_cost = calculate_order_total_price(orders_set)
print(f"\nTotal Price: **{total_cost:.2f}**")
````
